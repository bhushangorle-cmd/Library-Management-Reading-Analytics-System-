### 📚 Library Management Reading Analytics System

This project analyzes library borrowing data to identify patterns such as popular books, active readers, and late returns.

###  Problem Statement

Libraries store a large amount of data, but it is not effectively analyzed.

This project aims to:
- Identify popular books
- Track reader activity
- Detect late returns
- Analyze genre trends

### Technologies Used

- Python
- Object-Oriented Programming (OOP)
- CSV File Handling
- Collections Module
- Datetime Module

### Creating Classes

We define classes to structure the data:
- Book
- Member
- BorrowRecord

```python
### Load Data from CSV
import csv

books = {}
members = {}
records = []
```

###  Creating Classes

We define classes to structure the data:
- Book
- Member
- BorrowRecord

```python
### Classes
class Book:
    def __init__(self, book_id, title, author, genre):
        self.book_id = int(book_id)
        self.title = title
        self.author = author
        self.genre = genre


class Member:
    def __init__(self, member_id, name):
        self.member_id = int(member_id)
        self.name = name


class BorrowRecord:
    def __init__(self, record_id, member_id, book_id, borrow_date, return_date):
        self.record_id = int(record_id)
        self.member_id = int(member_id)
        self.book_id = int(book_id)
        self.borrow_date = borrow_date
        self.return_date = return_date
```

### 📂 Loading Data from CSV Files

Data is loaded from CSV files and stored in dictionaries and lists.

```python
### Load Books
with open('books.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        book = Book(**row)
        books[book.book_id] = book

print("Books Loaded:", len(books))
```

```python
### Load Members
with open('members.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        member = Member(**row)
        members[member.member_id] = member

print("Members Loaded:", len(members))
```

```python
### Load Borrow Records
with open('borrow_records.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        record = BorrowRecord(**row)
        records.append(record)

print("Records Loaded:", len(records))
```

###  Data Validation

In this step, we verify that all borrow records are valid.

We check:
- Whether the book ID exists in the books dataset
- Whether the member ID exists in the members dataset

Invalid records are ignored to ensure accurate analysis.

This step improves data reliability and prevents errors during processing.

```python
### Validate Data
valid_records = []

for record in records:
    if record.book_id in books and record.member_id in members:
        valid_records.append(record)
    else:
        print("Invalid Record:", record.record_id)

print("Valid Records:", len(valid_records))
```

### 📚 Most Borrowed Books Analysis

This section identifies the most frequently borrowed books in the library.

We calculate:
- The number of times each book has been borrowed

Using this analysis, we can:
- Determine popular books among readers
- Understand reading trends
- Help the library manage inventory efficiently

The result displays books ranked based on how often they are borrowed.

```python
### Find Most Borrowed Books
from collections import Counter

book_counter = Counter()

for record in valid_records:
    book_counter[record.book_id] += 1

for book_id, count in book_counter.most_common():
    print(books[book_id].title, "->", count)
```

### 👥 Active Readers Analysis

This section identifies the most active readers in the library.

We calculate:
- How many books each member has borrowed

Using this, we can:
- Find frequent users
- Understand reader engagement

This helps the library reward active readers or improve services.

```python
### Find Active Readers
member_counter = Counter()

for record in valid_records:
    member_counter[record.member_id] += 1

for member_id, count in member_counter.most_common():
    print(members[member_id].name, "->", count)
```

### ⏰ Late Return Detection

This step checks whether books are returned on time.

We calculate:
- The difference between borrow date and return date

If the number of days exceeds the allowed limit (e.g., 14 days),
the record is marked as a late return.

This helps in:
- Tracking delays
- Applying fines (if needed)

```python
### Find Late Returns
from datetime import datetime

LATE_DAYS = 14

for record in valid_records:
    borrow = datetime.strptime(record.borrow_date, "%Y-%m-%d")
    return_d = datetime.strptime(record.return_date, "%Y-%m-%d")

    days = (return_d - borrow).days

    if days > LATE_DAYS:
        print("Late Record:", record.record_id, "Late by", days - LATE_DAYS, "days")
```

### 📚 Popular Genre Analysis

In this section, we analyze which genres are most popular.

We calculate:
- How many times books from each genre are borrowed

This helps:
- Understand reader preferences
- Improve book collection planning

```python
### Find Popular Genres
genre_counter = Counter()

for record in valid_records:
    genre = books[record.book_id].genre
    genre_counter[genre] += 1

for genre, count in genre_counter.most_common():
    print(genre, "->", count)
```

### 📄 Report Generation

After analysis, the system generates a report file.

The report includes:
- Most borrowed books
- Active readers
- Other key insights

This report can be saved as a text file for future reference or decision-making.

```python
##Save Report
with open("report.txt", "w") as f:
    f.write("Library Report\n\n")

    for book_id, count in book_counter.items():
        f.write(books[book_id].title + " -> " + str(count) + "\n")

print("Report Saved!")
```
